# Funciones de distribución discreta y continua

Distribucion de probabilidad discreta

Propiedades

1. Rango
2. Probabilidad 
3. Probabilidad Acumulada
4. Regla de Completitud de Espacio
5. Valor Esperado (promedio)
6. Varianza (media - x observado)


```
# funciones de probabilidad

from scipy.stats import binom

p = 0.3 # probabilidad
n = 8 # numero de intentos

x = list(range(0,9,1)) 
y = list([])

for i in x:
  y.append(binom.pmf(i, p=p, n=n))

y
```

```
fig, ax = plt.subplots()
ax.bar(x, y)
ax.set_ylabel('probabilidad discreta')
np.sum(y)
```

```
media = np.average(x, weights = y)

varianza = np.average( ((x - media)**2), weights = y)

media, varianza

fig, ax = plt.subplots()
ax.bar(x, y)
ax.set_ylabel('probabilidad discreta')
ax.axvline(x=media, c = 'r', linestyle='--', label = "Valor Esperado")
ax.axvline(x=media+3*np.sqrt(varianza), c = 'k', linestyle="--", label = "Varianza")
ax.legend()
```

# Estimadores de Maxima Verosimilitud

El método de máxima verosimilitud consiste en obtener el valor de lambda donde la L(lambda) sea máximo.

L(lambda) es la función de máxima verosimilitud, y está definida como el producto entre todos los valores de la muestra aleatoria evaluados en su función de densidad.

Los estimadores de maxima verosimilitud seran funciones de la muestra que nos permitan calcular la version mejor del parametro estimado. 

En este caso la distribución a estudiar es una exponencial, esto es importante, porque TODA distribución exponencial tiene función de densidad = parametro * e^(parametro*x), entonces ya tenemos una función con que trabajar.

Se desarrolla la función para después aplicar Logaritmo natural.

¿Por qué logaritmo natural? Por dos razones, una tiene que ver con lo práctico de utilizar logaritmos en términos operatorios, pero la mas importante es que, dadas las propiedades de los logaritmos, **la función L(lambda) es máxima en el mismo punto que Ln(L(lambda)). **

Se desarrolla la nueva función que se simplifica gracias a las propiedades de los logaritmos.

Por qué se deriva y se iguala a cero? recuerden que la primera derivada hace referencia a la pendiente de la función, y si la pendiente es cero significa que estamos en presencia de un mínimo o un máximo.

Acá falto algo, porque para poder asegurar que ese valor de lambda es máximo, se debe derivar por segunda vez, si la segunda derivada es < 0 entonces estamos frente a un máximo, no se puede asegurar nada sin hacer este análisis.

Sorpresa! el estimador de max verosimilitud de una función es el promedio muestral.

El promedio muestral asegura:

* Lambda en una funcion exponencial y de Poisson
* P en Bernoulli
* Mu en una Normal

Y la Varianza Muestral sera el mejor estimador para las varianzas.

# Pruebas de hipotesis

Una prueba de hipotesis en una regla que especifica si se puede aceptar o rechazar una afirmacion acerca de un parametro poblacional  (lambda, sigma, mu, etc) dependiendo de la evidencia proporcionada por una muestra de datos


1. Identificar la parametro poblacional sobre el que queremos hacer inferencias 

2. Hipotesis Nula (Siempre igualdad) e Hipotesis Alterna (mayor que >, menor que, diferente de)

# Intervalos de confianza


Los IC nos permiten encontrar un limite inferior y superior, tales que acumulen una cierta probabilidad o confianza, de que el parametro poblacional estara contenido y podra tomar ese valor en dif escenarios bajo los cuales sea medida esa variable

Un IC es un rango de valores, derivado de estadisticos de la muestra, que con cierto grado de confianza, incluye el valor de una parametro poblacional desconocido

Un IC de 95% indica que 19 de 20 muestras de la misma poblacion produciran IC que contendran el parametro de poblacion

¿Como calcular este IC asociado a una variable aleatoria?


1. Definir el parametro a probar (teta, lambda, etc)

2. Identificar el estimador (promedio muestral o varianza muestral)

3. Identificar el estadistico y su distribucion (norm, t, chi, f)

4. Tolerancia al error alfa, estimacion puntual y rango de error

5. Conclusion del IC

6. Opcional: Pruebas de hipotesis usando IC

# Regresion Logistica

Regresion Lineal --- variable de respuesta cuantitativa

Regresion Logistica -- Variables de respuesta cualitativa 

Variable cualitativa -- Naturaleza Bernoulli 

X = 0 No comprador

X = 1 Comprador

La regresion lineal no se puede aplicar porque al estimar los valores de 0 a 1 incurririamos en valores menores a 0 y mayores que 1

Tranformamos la variable de salida con el operador logistico, el cual acotara el resultado a 0 o a 1, a traves de la siguiente formula:

ProbL = 1 / 1 + e^-(B0+B1x1+...Bkxk)
